```python
import pandas as pd
import matplotlib.pyplot as plt

def show_bar_graph(filename, name):
    df = pd.read_csv(filename)
    df[name].value_counts().plot(kind='bar', color='black')
    plt.title(f'Bar Graph for {name}')
    plt.xlabel(name)
    plt.ylabel('')
    plt.show()

filename = "cars.csv"
column_to_plot = "brand"
show_bar_graph(filename, column_to_plot)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def show_pie_chart_with_other(filename, name, threshold=0.05):
    df = pd.read_csv(filename)
    value_counts = df[name].value_counts()
    small_brands = value_counts[value_counts / len(df) < threshold].index
    df[name] = df[name].replace(small_brands, 'Other')
    plt.figure(figsize=(8, 8))
    df[name].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap='viridis')
    plt.title(f'Pie Chart for {name}')
    plt.ylabel('') 
    plt.show()

filename = "cars.csv"
column_to_plot = "brand"
show_pie_chart_with_other(filename, column_to_plot)
```
